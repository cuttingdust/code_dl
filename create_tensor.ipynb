```python
# 6 创建指定值的张量
t14 = torch.randn(size=(2,3,4))
t14
```

```python
"""
    1- ones创建指定形状的全1张量
    2- ones_like，仿照某个张量的形状，创建指定形状的全1张量
    3- zeros创建指定形状的全0张量。【掌握】
    4- zeros_like，仿照某个张量的形状，创建指定形状的全0张量
    5- full创建指定形状，并且元素值也是人为指定的张量
    6- full_like，仿照某个张量的形状，并且元素值是人为指定的张量
"""
t15 = torch.ones(size=(3,4))
t16 = torch.ones_like(t14)
t17 = torch.zeros(size=(3,4))
t18 = torch.zeros_like(t14)
t19 = torch.full(size=(3,4), fill_value=999)
t20 = torch.full_like(t14, 666)

print(f"ones-->{t15}")
print(f"ones_like-->{t16}")
print(f"zeros-->{t17}")
print(f"full_like-->{t18}")
print(f"full_like-->{t19}")
print(f"full_like-->{t20}")
```

```python
# 7 张量元素数据类型转换
t_1 = torch.tensor([11, 22, 33])
t_1.dtype
```

```python
# 通用的写法。掌握
t_2 = t_1.type(dtype=torch.float16) # 推荐的写法
print(t_2)
t_2.dtype
```

```python
t_2 = t_1.type(dtype=torch.DoubleTensor)
print(t_2)
t_2.dtype
```

```python
t_3 = torch.tensor([1.678, 2.543])  # 默认是float32])
print(t_3)
t_3.dtype
```

```python
t_4 = t_3.half()   # float16
t_4.dtype
```

```python
t_4 = t_3.short()  # 短整型int16
t_4.dtype
```

```python
t_4 = t_3.int()  # int32
t_4.dtype
```

```python
t_4 = t_3.long()   # 长整型int64
t_4.dtype
```
